```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/UNSW_2018_IoT_Botnet_Full5pc_2.csv')

# Basic info
print(df.shape)
df.head()
```

```python
attack_counts = df['attack'].value_counts()

print(attack_counts)

plt.figure()
attack_counts.plot(kind='bar')
plt.title('Attack vs Normal Distribution')
plt.xlabel('Attack Label')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,4))
df['category'].value_counts().plot(kind='bar')
plt.title('Attack Categories Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,4))
df['subcategory'].value_counts().plot(kind='bar')
plt.title('Attack Subcategories Distribution')
plt.show()
```
